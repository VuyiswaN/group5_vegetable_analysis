Since Methi shows an outlier and is a numerical (continous) variable we will replace it with the mean


```python
def channge_outlier(x):
    if x >= 2000:
        return df['Methi'].mean()
    else:
        return x

df['Methi'] = df['Methi'].astype(float)
df['Methi'] = df['Methi'].apply(channge_outlier)

df.describe()
```

# 5. Exploratory Data Analysis (EDA)

EDA is a crucial step in the data analysis pipeline. It is the process of analyzing and visualizing datasets to summarize their main characteristics and extract useful insights, often with the help of statistical graphics, plots, and other data visualization tools.


# Univariate Analysis
#### KDE plots for each vegetable price distribution 


```python
numerical_columns = df.select_dtypes('number')
# Assuming MD_agric_df is already defined with the simulated data

# Setting up a 5x4 grid of plots
fig, axes = plt.subplots(3, 4, figsize=(15, 15))  # Adjust the figure size as needed
axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

# Plotting a KDE for each column in its respective subplot
for i, column in enumerate(numerical_columns.columns):
    mean_val = numerical_columns[column].mean()
    sns.kdeplot(numerical_columns[column], ax= axes[i])
    axes[i].set_title(f"{column} Price KDE")
    axes[i].axvline(mean_val, color='red', linestyle='dashed', linewidth=2)


plt.tight_layout()
plt.show()
```

The above kde plots show price distribution around the mean.

Onion, brinjal, methi and green chili distributions are slightly left-skewed, indicating that most values fall below the mean. As a result, the mean may not be the best measure of central tendency, as it is influenced by higher, maximal values

The distribution of potato values is right-skewed, meaning most values are above the mean. Consequently, the mean may not be the best measure of central tendency, as it is affected by extremely low values

Tomato are evenly distributed around the mean, indicating that the data is balanced around the center. In this case, the mode, median, and mean are fairly similar or close to each other, providing a consistent measure of central tendency

Garlic, Suran and Peas are multi- modal, this means they have different peak prices. Whereas the rest have single price peaks.
In this case, the median and mode are often better indicators of central tendency than the mean

# Hypothesis Testing

Our aim is to test difference in means across different vegetables price and to test normality on each distribution

Null Hypothesis: 
$$
H_0: \mu 1 = \mu 2
$$
Alternate Hypothesis:
$$
H_1: \mu 1 \neq \mu 2
$$


We are going to use the Anova which allows us to test in more two groups


```python
df_melted = pd.melt(df, value_vars=['Tomato', 'Onion', 'Potato','Bhindi (Ladies finger)','Elephant Yam (Suran)'], 
                    var_name='Vegetable', value_name='Price')
```

This function unpivots the data meaning it transforms the columns into rows.In this scenario unpivoting is neccesaary in odrer to apply the anova function.
The anova function requires both independent and dependent columns and that was perfectly attained after implemeting this code.

```python
Tomato = df_melted['Price'][df_melted['Vegetable'] == 'Tomato']
Onion= df_melted['Price'][df_melted['Vegetable'] == 'Onion']
Potato = df_melted['Price'][df_melted['Vegetable'] == 'Potato']
Bhindi= df_melted['Price'][df_melted['Vegetable'] == 'Bhindi (Ladies finger)']
Suran= df_melted['Price'][df_melted['Vegetable'] =='Elephant Yam (Suran)']

f_stat, p_value = stats.f_oneway(Tomato, Onion, Potato)
f_stat2, p_value2=stats.f_oneway(Bhindi,Suran)

print(f"F-statistic for Tomato,Onion and Potato: {f_stat}")
print(f"P-value: {p_value}")

print(f"F-statistic for Bhindi and Suran: {f_stat2}")
print(f"P-value: {p_value2}")

```

We are specifically testing Toamto,Onion and Potato because they have similar means which may imply their population averages are equal.
Bhindi and Suran also have similar avearges.

The pvalue is extreemely small meaning it is safe to reject the null hypothesis,there is a difference in Tomato,Potato and Onion vegetable price.
The 2nd pvalue is bigger than 0.05 meaning that we cannot reject the the null hyphothesis, there is no difference in Bhindi and Suran population mean


# Conclusion

In conclusion, vegetable price fluctuations result from various factors, including farming seasons, availability, weather conditions (droughts, heatwaves, floods), demand, and transportation logistics. Among the vegetables analyzed, garlic, peas, and green chili have the highest average prices.

Garlic shows a consistent upward price trend throughout the year, while peas exhibit seasonal peaks every three months. Green chili maintains a stable price trend with a notable peak in month seven. Conversely, tomatoes, bhindi (okra), elephant yam, and potatoes display stationary prices with no significant trends. Onions reflect a slight upward trend over the year.

These findings highlight the significant role of seasonality and demand in vegetable pricing. Garlic and onions are expected to see continuous price increases due to their essential culinary roles and health benefits, making them potential investment opportunities.
During harvest seasons, prices typically drop due to oversupply; however, off-season production often relies on greenhouse farming, leading to limited supply and increased demand that drives prices up. Peas are primarily planted in cooler seasons like spring but can also be grown in winter, contributing to their price fluctuations. In contrast, tomatoes, bhindi, elephant yam, and potatoes maintain stable prices due to balanced supply and demand.
